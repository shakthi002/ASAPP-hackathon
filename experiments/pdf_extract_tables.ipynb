```python
import os
import pandas as pd
from tabula import read_pdf
from groq import Groq  # Ensure this library is installed
import pinecone
from sentence_transformers import SentenceTransformer

# Initialize Pinecone
pc = pinecone.Pinecone(api_key='082f4c9e-e92c-4012-a021-17bafd9cffea', environment='us-east-1')

# Create a Pinecone index if it doesn't exist
index_name = "table-index3"
if index_name not in pc.list_indexes().names():
    pc.create_index(index_name)

# Connect to the index
index = pc.Index(index_name)

# Initialize SentenceTransformer model for embeddings
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')  # You can choose a different model if needed

GROQ_API_KEY = "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"

def extract_tables_from_pdf(pdf_path):
    """Extract tables from a single PDF file using Tabula."""
    try:
        tables = read_pdf(pdf_path, pages='all', multiple_tables=True)
        return tables  # Returns a list of DataFrames
    except Exception as e:
        print(f"Error extracting tables from {pdf_path}: {e}")
        return []

def extract_tables_from_multiple_pdfs(pdf_paths):
    """Extract tables from multiple PDF files."""
    all_tables = {}

    for pdf_path in pdf_paths:
        if os.path.exists(pdf_path):
            tables = extract_tables_from_pdf(pdf_path)
            all_tables[pdf_path] = tables
        else:
            print(f"File not found: {pdf_path}")

    return all_tables

def summarize_table_with_llm(table_text):
    """Summarize the table data using the Groq Llama3 model."""
    client = Groq(api_key=GROQ_API_KEY)
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": f"Summarize the following table data:\n{table_text}",
            }
        ],
        model="llama3-8b-8192",
    )

    summary = chat_completion.choices[0].message.content  # Accessing summary correctly
    return summary

def generate_embedding(text):
    """Generate embeddings using SentenceTransformer."""
    embedding = embedding_model.encode(text).tolist()  # Convert to list for Pinecone compatibility
    return embedding

def pass_to_llm_and_store_in_pinecone(tables):
    """Pass extracted tables to the Groq Llama3 model for summarization and store in Pinecone."""
    summaries = []

    for pdf_file, tables in tables.items():
        for i, table in enumerate(tables):
            table_text = table.to_string(index=False)  # Convert DataFrame to string
            summary = summarize_table_with_llm(table_text)
            summaries.append({"pdf_file": pdf_file, "table_index": i, "summary": summary})

            # Generate embedding for the summary
            embedding = generate_embedding(summary)

            # Store in Pinecone (using PDF filename and table index as ID)
            index.upsert([(f"{pdf_file}_{i}", embedding)])

    return summaries

# Example usage
pdf_files = ['sample1.pdf', 'sample2.pdf']  # Add your PDF file paths here
extracted_tables = extract_tables_from_multiple_pdfs(pdf_files)
summaries = pass_to_llm_and_store_in_pinecone(extracted_tables)

# Optional: Print stored summaries
for summary in summaries:
    print(f"Stored Summary for {summary['pdf_file']} Table {summary['table_index']}:\n{summary['summary']}\n")
```
